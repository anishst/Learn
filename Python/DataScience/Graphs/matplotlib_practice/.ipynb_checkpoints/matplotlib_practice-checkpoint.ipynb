```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,4,5]
y = [2,3,5,3]

plt.plot(x,y)
plt.title('First graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x,y)
plt.title('First graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('x axis')
plt.ylabel('Y axis')

# change tick values
plt.xticks([0,1,2,3])
plt.yticks([1,2,3])
plt.show()
```
