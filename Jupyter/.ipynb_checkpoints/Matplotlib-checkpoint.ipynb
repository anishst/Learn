# Simple Plot

```python
from matplotlib import pyplot as plt

# plotting to our canvas
plt.plot([1,2,3], [4,5,1])
# show the plot
plt.show() 
```

# Using variables and axis title and labels

```python
from matplotlib import pyplot as plt

x = [5,8,10]
y = [12,16,6]

plt.plot(x,y)
# set title
plt.title("Graph Title")
plt.ylabel("y axis")
plt.xlabel("x axis")

plt.show()

```

# Adding Styles to graph

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")


x = [5,8,10]
y = [12,16,6]

x2 = [6,8,10]
y2 = [6,15,4]

plt.plot(x,y,'b', label="LineOne", linewidth=5)
plt.plot(x2,y2,'c', label="LineTwo", linewidth=5)

# set title
plt.title("Graph Title")
plt.ylabel("y axis")
plt.xlabel("x axis")

plt.legend()

plt.grid(True,color='k')

plt.show()
```

# How to plot bar graph

```python
from matplotlib import pyplot as plt

plt.bar([1,6,3,5],[4,5,6,7], label="example one")
plt.bar([3,4,3,5],[14,15,12,17], label="example two", color="g")
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title("Bar graph")

plt.show()
```
